### Advance operation with Bussiness examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data structure :[restaurant_id ,2021,2022,2023,2024]
sales_data=np.array([
    [1, 150000, 180000, 220000, 250000],#paradise
    [2, 120000, 140000, 160000, 190000],#beijing Bites
    [3, 200000, 230000, 260000, 300000],#Pizza Point
    [4, 180000, 210000, 240000, 270000],#Burger Point
    [5, 160000, 185000, 205000, 230000] #Chai point
])

print("=== Zomato Sales Analysis ===")
print("\n Sales data shape",sales_data.shape)
print("\n Sample data for first three retaurant:",sales_data[0:3]) #slice
```

```python
#total sales per year
print(np.sum(sales_data,axis=0)) # vertical sum up to down
yearly_total=np.sum(sales_data[:,1:],axis=0) # now excluding sr.no
print(yearly_total)

#minimum sales per restaurant
min_sales=np.min(sales_data[:,1:],axis=1)
print("minium sales :\n",min_sales)

#maximum sales per year
max_sales=np.max(sales_data[:,1:],axis=0)
print("max sales:",max_sales)

#average sales per restaurant 
avg_sales=np.average(sales_data[:,1:],axis=1)
print("avg sales of restaurant :",avg_sales)
```

```python
#cumulative sales : first year sales than adding another year and than adding another year 
cumsum1=np.cumsum(sales_data[:,1:],axis=1)
print("cumulative sales:",cumsum1)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum1,axis=0))
plt.title("Average cumulative sales across all restaurant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
